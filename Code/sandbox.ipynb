```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff

data = pd.read_csv('data.csv')
data.head() #The head() function is used to get the first n rows. The default value is 5 aka data.head(self,n=5)
```

```python
data.describe()
```

```python
# data = data.drop(['Unnamed: 32','id'],axis = 1) # This line has been commented out as it has already been run once and it will throw a "doesn't exixt error"
data.diagnosis.replace(to_replace = dict(M = 1, B = 0), inplace = True)
```

```python
data.describe()
```

```python
data.head()
```

```python
def noErrors():
    print("Compilation has been completed and there are no errors in the code.")
    return True
```

```python
M = data[(data['diagnosis'] != 0)]
B = data[(data['diagnosis'] == 0)]
```

```python
trace = go.Bar(x = (len(M), len(B)), y = ['Malignant', 'Benign'], orientation = 'h', opacity = 0.7, marker=dict(
        color=[ 'red', 'green'],
        line=dict(color='#000000',width=1.5)))

layout = dict(title =  'Pictorial count of the data available in the dataset.')
                    
fig = dict(data = [trace], layout=layout)
py.iplot(fig)
noErrors()
```

```python
def plot_distribution(data_select, size_bin) :  
    tmp1 = M[data_select]
    tmp2 = B[data_select]
    hist_data = [tmp1, tmp2]
    
    group_labels = ['malignant', 'benign']
    colors = ['red', 'green']

    fig = ff.create_distplot(hist_data, group_labels, colors = colors, show_hist = True, bin_size = size_bin, curve_type='kde')
    
    fig['layout'].update(title = data_select)

    py.iplot(fig, filename = 'Data plot representation')
```

```python
plot_distribution('radius_mean', .5)
```

```python
plot_distribution('texture_mean', .5)
```

```python
plot_distribution('perimeter_mean', 5)
```

```python
plot_distribution('area_mean', 10)
```
